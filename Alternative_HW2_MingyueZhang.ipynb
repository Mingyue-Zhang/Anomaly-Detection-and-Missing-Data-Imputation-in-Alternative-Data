### Validation
Procedure to determine the number of neighbors used in the KNN Imputer:

Step 1: randomly delete data in the dataset based on the distribution of the size of missing data chunks unitl the percentage of missing data reaches predetermined value (achieved by the function "delete")

Generate N datasets by doing so.

Step 2: get the performance of each imputer on all the N datasets we got from step1.(achieved by the funciton "imputer_validate"）

Step 3: get the optimal imputer

```python
#validate
import random
import math
from sklearn.impute import KNNImputer

#function to randomly delete data based on the distribution of missing data
def delete(data, goal=0.1):
    """
    data: array
    goal: fraction of missing data in the whole dataset
    """
    totalNumber = data.shape[0]*data.shape[1]
    existingMissing = sum(sum(np.isnan(data)))
    goalNumber = math.ceil(goal*totalNumber)
    
    new_data = data.copy()
    
    if existingMissing >= goalNumber:
        print('Original data already has',str(existingMissing/totalNumber*100),
              '% missing data, no need to delete more.')
        return new_data
    
    #print(existingMissing/totalNumber*100)
    monthNumber =  len(data[0])
    missing = np.isnan(data).sum(axis=1)
    missing_distribution = missing[missing>0]
    average_missing_size = round(sum(missing)/(np.count_nonzero(missing)))                                 
    
    #chunk_size = random.choices(missing_distribution, np.ceil(goalNumber-existingMissing))
    
    while(existingMissing - goalNumber<-totalNumber*0.005):
        idx = random.sample(range(len(data)), k=math.ceil((goalNumber-existingMissing)/average_missing_size))
        #print(idx)
        missing = np.isnan(new_data).sum(axis=1)                           

        for i in idx:   
            if missing[i]+average_missing_size>=monthNumber:
                new_data[i] = [np.nan]*monthNumber
            else:
                month = random.sample(range(monthNumber-average_missing_size),k=1)[0]
                new_data[i][month:month+average_missing_size] = np.nan
    
        existingMissing = sum(sum(np.isnan(new_data)))
                                                      
    return new_data
```

```python
from sklearn.metrics.pairwise import nan_euclidean_distances

#function to validate the parameters of imputer
def imputer_validate(originalData, params, validationNumber=5, goal=0.1):
    """
    params: list of candidate parameters
    validationNumber: number of validation dataset to generate
    """
    validateDataList = []
    for i in range(validationNumber):
        validateDataList.append(delete(originalData,goal))
    
    distance_stats = []
    
    for param in params:
        imputer = KNNImputer(n_neighbors=param,weights='distance')
        distance = 0
        for data in validateDataList:
            imputation = imputer.fit_transform(data)
            distance+=np.nansum(np.diag(nan_euclidean_distances(data, imputation)))
        distance_stats.append(distance)
        #print(distance)

    return params[np.argmin(distance_stats)]
            
```

```python
candidate_params = range(1,31)

best_param = imputer_validate(df2.values, candidate_params, validationNumber=10, goal=0.3)
```

### Hot Deck Imputation

A randomly chosen value from an individual in the sample who has similar values on other variables.

In other words, find all the sample subjects who are similar on other variables, then randomly choose one of their values on the missing variable.
